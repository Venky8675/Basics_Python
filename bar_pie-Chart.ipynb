```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = pd.DataFrame({
    'Cars': ['Audi', 'BMW', 'Mercedus'],
    'Value': [124, 98, 113],
    
})

# Set up the plot
plt.figure(figsize=(10, 6))

# Bar plot
sns.barplot(data=data, x='Cars', y='Value',alpha=0.7)

# Line plot
sns.lineplot(data=data, x='Cars', y='Value', marker='o')

# Add title and labels
plt.title('Bar and Line Plot for Single Dataset')
plt.xlabel('Cars')
plt.ylabel('Value')

# Show legend
plt.legend()

# Display plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import PercentFormatter

# Sample data: causes and their frequencies
cars = ['Audi', 'Mercedes', 'BMW']
frequency = [124,113,98]

# Create a DataFrame and sort by frequency in descending order
df = pd.DataFrame({'cars': cars, 'Frequency': frequency})
df = df.sort_values(by='Frequency', ascending=False)
df['Cumulative Percentage'] = df['Frequency'].cumsum() / df['Frequency'].sum() * 100

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for frequencies
ax.bar(df['cars'], df['Frequency'], color='C0', label='Frequency')

# Line plot for cumulative percentage
ax2 = ax.twinx()
ax2.plot(df['cars'], df['Cumulative Percentage'], color='C1', marker='o', linestyle='-', label='Cumulative %')
ax2.yaxis.set_major_formatter(PercentFormatter())

# Labels and titles
ax.set_xlabel('Cars')
ax.set_ylabel('Frequency')
ax2.set_ylabel('Cumulative Percentage')
plt.title('Pareto Diagram')

# Show legend
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
